# Computación Gráfica y Visual
Utilización de librerías Numpy y Matplotlib para la visualización de gráficas y formas geometricas.

```python
#Area
import numpy as np
import matplotlib.pyplot as plt

x1=-10
x2=10
y1=-10
y2=10
plt.axis([x1,x2,y1,y2])

plt.axis('on')
plt.grid(True)

plt.show()
```

```python
# Area
x1=-10
x2=140
y1=90
y2=-10
plt.axis([x1,x2,y1,y2])
plt.axis('on')

#grid
plt.grid(True,color='b')
plt.title('Ejemplo Marcadores')

#Marcadores
xmin=x1
xmax=x2
dx=10
ymin=y1
ymax=y2
dy=-5
plt.xticks(np.arange(xmin, xmax, dx))
plt.yticks(np.arange(ymin, ymax, dy))

plt.show()
```

```python
import matplotlib.pyplot as plt
#Puntos
x1=-5
x2=15
y1=-15
y2=5

plt.axis([x1,x2,y1,y2])

plt.axis('on')

dx=.5
dy=.5
for x in np.arange(x1,x2,dx):
    for y in np.arange(y1,y2,dy):
        plt.scatter(x,y,s=1,color='grey')

plt.show()
```

```python

#Degradado de colores
plt.axis([0,100,0,10])

for x in np.arange(1,100,1):
    r=0
    g=0
    b=x/100
    plt.plot([x,x],[0,10],linewidth=5,color=(r,g,b))

plt.show()
```

```python
#Atenuaciones
plt.axis([0,150,0,100])

plt.scatter(60,50,s=1000,color='b',alpha=1)
plt.scatter(80,50,s=1000,color='b',alpha=.5)
plt.scatter(100,50,s=1000,color='b',alpha=.1)

plt.show()
```

```python
plt.grid(True)
plt.axis('on')

plt.axis([-10,10,10,-10])

#grid
x1=-10
x2=10
y1=10
y2=-10

dx=.5
dy=-.5
for x in np.arange(x1,x2,dx):
    for y in np.arange(y1,y2,dy):
        plt.scatter(x,y,s=1,color='blue')

#Cuadrado
plt.plot([-5,5],[-5,-5],linewidth=2,color='k')
plt.plot([5,5],[-5,5],linewidth=2,color='k')
plt.plot([5,-5],[5,5],linewidth=2,color='k')
plt.plot([-5,-5],[5,-5],linewidth=2,color='k')

plt.show()
```

```python
# Ovalo
plt.axis([0,100,100,0])

r=40
alpha1=np.radians(0)
alpha2=np.radians(360)
dalpha=np.radians(2)
xc=50
yc=50
plt.scatter(xc,yc,s=10,color='k')
for alpha in np.arange(alpha1,alpha2,dalpha):
    x=xc+r*np.cos(alpha)
    y=yc+r*np.sin(alpha)
    plt.scatter(x,y,s=5,color='k')

plt.show()
```
